```python
def perfil_jugadores(df_eventos):
    # Eliminar eventos sin jugador
    df = df_eventos.dropna(subset=['player_id', 'player']).copy()

    # Lista para guardar los resultados
    perfiles = []

    # Agrupar por jugador
    for player_id, grupo in df.groupby('player_id'):
        nombre = grupo['player'].iloc[-1]
        ultimo_equipo = grupo['team'].dropna().iloc[-1] if 'team' in grupo else None
        ultimo_team_id = grupo['team_id'].dropna().iloc[-1] if 'team_id' in grupo else None
        posiciones = grupo['position'].dropna().unique().tolist()
        partidos_jugados = grupo['match_id'].nunique()  # contar partidos únicos

        perfiles.append({
            'player_id': int(player_id),
            'player': nombre,
            'ultimo_equipo': ultimo_equipo,
            'ultimo_team_id': ultimo_team_id,
            'posiciones': posiciones,
            'partidos_jugados': partidos_jugados
        })

    return pd.DataFrame(perfiles)


perfiles_jugadores = perfil_jugadores(eventos_bundesliga_16)
```

```python
from rapidfuzz import process, fuzz
import pandas as pd

def hacer_matching_aproximado(base_df, columna_base, df_referencia, columna_referencia, umbral=85):
    """
    Realiza un emparejamiento difuso (fuzzy matching) entre dos listas de nombres.

    Parámetros:
    - base_df: DataFrame con los nombres originales (por ejemplo, nombres en statsbomb).
    - columna_base: columna en base_df con los nombres a emparejar (ej: 'team', 'player').
    - df_referencia: DataFrame con los nombres de referencia (por ejemplo, datos económicos).
    - columna_referencia: columna en df_referencia con los nombres correctos (ej: 'team').
    - umbral: valor mínimo de similitud para aceptar automáticamente (default = 85).

    Devuelve:
    - DataFrame con columnas: nombre original, nombre emparejado, score, revisar (True/False)
    """
    nombres_base = base_df[columna_base].unique()
    nombres_ref = df_referencia[columna_referencia].tolist()

    resultados = []

    for nombre in nombres_base:
        mejor_match, score, _ = process.extractOne(nombre, nombres_ref, scorer=fuzz.token_sort_ratio)
        revisar = score < umbral
        resultados.append({
            columna_base: nombre,
            f"{columna_referencia}_match": mejor_match,
            "score": score,
            "revisar": revisar
        })

    return pd.DataFrame(resultados)
```

```python
import pandas as pd
from rapidfuzz import process, fuzz

# 1. Cargar datasets ya cargados:
# df_jugadores
# df_jugadores_fifa_laliga

# 2. Fuzzy matching entre nombres de jugadores
def hacer_matching_aproximado(base_df, columna_base, df_referencia, columna_referencia, umbral=85):
    nombres_base = base_df[columna_base].unique()
    nombres_ref = df_referencia[columna_referencia].tolist()

    resultados = []

    for nombre in nombres_base:
        resultado = process.extractOne(nombre, nombres_ref, scorer=fuzz.token_sort_ratio)
        if resultado:
            mejor_match, score, _ = resultado
            revisar = score < umbral
        else:
            mejor_match, score, revisar = None, 0, True

        resultados.append({
            columna_base: nombre,
            f"{columna_referencia}_match": mejor_match,
            "score": score,
            "revisar": revisar
        })

    return pd.DataFrame(resultados)

# 3. Aplicar fuzzy matching jugador ↔ nombre FIFA
df_matching = hacer_matching_aproximado(
    base_df=df_jugadores,
    columna_base='player',
    df_referencia=df_jugadores_fifa_laliga,
    columna_referencia='long_name',
    umbral=85
)

# 4. Merge para traer el match
jugadores_matcheados = df_jugadores.merge(df_matching, on='player')

# 5. Merge con FIFA para traer info de jugador
jugadores_completo = jugadores_matcheados.merge(
    df_jugadores_fifa_laliga,
    left_on='long_name_match',
    right_on='long_name',
    how='left',
    suffixes=('', '_fifa')
)

# 6. Filtrar para que club_name (de FIFA) coincida con ultimo_equipo (StatsBomb)
jugadores_completo = jugadores_completo[jugadores_completo['ultimo_equipo'] == jugadores_completo['club_name']]

# 7. Guardar resultado final
jugadores_completo.to_csv("jugadores_con_datos_fifa_bundesliga.csv", index=False)
print("✅ Archivo guardado como 'jugadores_con_datos_fifa_bundesliga.csv'")

# 8. Mostrar jugadores que no se emparejaron bien
no_encontrados = jugadores_matcheados[~jugadores_matcheados['player'].isin(jugadores_completo['player'])]
print("\n❗ Jugadores que no se emparejaron con el mismo equipo:")
print(no_encontrados[['player', 'long_name_match', 'score', 'revisar']])

```
