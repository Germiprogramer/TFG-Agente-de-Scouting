```python
def perfil_jugadores(df_eventos):
    # Eliminar eventos sin jugador
    df = df_eventos.dropna(subset=['player_id', 'player']).copy()

    # Lista para guardar los resultados
    perfiles = []

    # Agrupar por jugador
    for player_id, grupo in df.groupby('player_id'):
        nombre = grupo['player'].iloc[-1]
        ultimo_equipo = grupo['team'].dropna().iloc[-1] if 'team' in grupo else None
        ultimo_team_id = grupo['team_id'].dropna().iloc[-1] if 'team_id' in grupo else None
        posiciones = grupo['position'].dropna().unique().tolist()
        partidos_jugados = grupo['match_id'].nunique()  # contar partidos únicos

        perfiles.append({
            'player_id': int(player_id),
            'player': nombre,
            'ultimo_equipo': ultimo_equipo,
            'ultimo_team_id': ultimo_team_id,
            'posiciones': posiciones,
            'partidos_jugados': partidos_jugados
        })

    return pd.DataFrame(perfiles)


perfiles_jugadores = perfil_jugadores(eventos_liga)
```

```python
from rapidfuzz import process, fuzz
import pandas as pd

def hacer_matching_aproximado(base_df, columna_base, df_referencia, columna_referencia, umbral=85):
    """
    Realiza un emparejamiento difuso (fuzzy matching) entre dos listas de nombres.

    Parámetros:
    - base_df: DataFrame con los nombres originales (por ejemplo, nombres en statsbomb).
    - columna_base: columna en base_df con los nombres a emparejar (ej: 'team', 'player').
    - df_referencia: DataFrame con los nombres de referencia (por ejemplo, datos económicos).
    - columna_referencia: columna en df_referencia con los nombres correctos (ej: 'team').
    - umbral: valor mínimo de similitud para aceptar automáticamente (default = 85).

    Devuelve:
    - DataFrame con columnas: nombre original, nombre emparejado, score, revisar (True/False)
    """
    nombres_base = base_df[columna_base].unique()
    nombres_ref = df_referencia[columna_referencia].tolist()

    resultados = []

    for nombre in nombres_base:
        mejor_match, score, _ = process.extractOne(nombre, nombres_ref, scorer=fuzz.token_sort_ratio)
        revisar = score < umbral
        resultados.append({
            columna_base: nombre,
            f"{columna_referencia}_match": mejor_match,
            "score": score,
            "revisar": revisar
        })

    return pd.DataFrame(resultados)
```

```python
import pandas as pd
from rapidfuzz import process, fuzz

def hacer_matching_aproximado_por_equipo(df_jugadores, df_fifa, umbral=85):
    """
    Empareja jugadores de StatsBomb con sus homólogos en FIFA buscando solo dentro del mismo equipo.
    Si no hay match en ese equipo, se busca globalmente y se marca equipo_match=False.
    """
    resultados = []

    for _, row in df_jugadores.iterrows():
        nombre_jugador = row['player']
        equipo_jugador = row['ultimo_equipo']

        # Jugadores FIFA del mismo equipo
        fifa_mismo_equipo = df_fifa[df_fifa['club_name'] == equipo_jugador]

        if not fifa_mismo_equipo.empty:
            nombres_candidatos = fifa_mismo_equipo['long_name'].tolist()
            scope_equipo = True
        else:
            nombres_candidatos = df_fifa['long_name'].tolist()
            scope_equipo = False  # No hay jugadores de ese equipo en FIFA

        mejor_match, score, _ = process.extractOne(
            nombre_jugador,
            nombres_candidatos,
            scorer=fuzz.token_sort_ratio
        ) or (None, 0, None)

        resultados.append({
            'player': nombre_jugador,
            'long_name_match': mejor_match,
            'score': score,
            'revisar': score < umbral,
            'equipo_match': scope_equipo
        })

    return pd.DataFrame(resultados)


# 🧪 Aplicación del proceso
df_matching = hacer_matching_aproximado_por_equipo(df_jugadores, df_jugadores_fifa_liga, umbral=65)

# Merge de resultado de matching con los jugadores originales
jugadores_matcheados = df_jugadores.merge(df_matching, on='player')

# Merge para traer datos de FIFA del jugador matcheado
jugadores_completo = jugadores_matcheados.merge(
    df_jugadores_fifa_liga,
    left_on='long_name_match',
    right_on='long_name',
    how='left',
    suffixes=('', '_fifa')
)

# Guardar CSV final
jugadores_completo.to_csv("jugadores_con_datos_fifa_laliga.csv", index=False)
print("✅ Archivo guardado como 'jugadores_con_datos_fifa_laliga.csv'")

# Mostrar jugadores donde no coincidió el equipo
no_match_equipo = jugadores_completo[jugadores_completo['equipo_match'] == False]
print("\n❗ Jugadores sin match de equipo (matching global):")
print(no_match_equipo[['player', 'long_name_match', 'score', 'club_name', 'ultimo_equipo']])
```
