```python
#lo hago solo de una liga porque sino me explota el ordenador
def obtener_eventos_por_partidos(lista_match_ids):
    """
    Descarga y concatena todos los eventos de los partidos dados.
    """
    eventos_todos = []

    for match_id in lista_match_ids:
        try:
            match_id_int = int(match_id)
            eventos = sb.events(match_id=match_id_int)
            eventos['match_id'] = match_id
            eventos_todos.append(eventos)
        except Exception as e:
            print(f"❌ Error en partido {match_id}: {e}")
            continue

    return pd.concat(eventos_todos, ignore_index=True) if eventos_todos else pd.DataFrame()

datos = competicion

match_ids = datos['match_id'].tolist()

df_eventos = obtener_eventos_por_partidos(match_ids)
```

```python
import os
import time
import pandas as pd
from statsbombpy import sb

def guardar_eventos_por_partido(match_ids, carpeta_salida="eventos_guardados"):
    os.makedirs(carpeta_salida, exist_ok=True)
    guardados = 0

    for i, match_id in enumerate(match_ids, 1):
        ruta_archivo = os.path.join(carpeta_salida, f"match_{match_id}.pkl")

        if os.path.exists(ruta_archivo):
            continue  # Ya descargado

        try:
            eventos = sb.events(match_id=int(match_id))
            eventos['match_id'] = match_id
            eventos.to_pickle(ruta_archivo)
            guardados += 1
        except Exception as e:
            print(f"❌ Error con partido {match_id}: {e}")
            continue

        if i % 20 == 0 or i == len(match_ids):
            total_guardados = len(os.listdir(carpeta_salida))
            print(f"📦 Guardados {total_guardados} partidos hasta ahora")

        time.sleep(1)  # Espera para no saturar

    print(f"\n✅ Proceso terminado. Total partidos guardados: {len(os.listdir(carpeta_salida))}")
```
