```python
#lo hago solo de una liga porque sino me explota el ordenador
def obtener_eventos_por_partidos(lista_match_ids):
    """
    Descarga y concatena todos los eventos de los partidos dados.
    """
    eventos_todos = []

    for match_id in lista_match_ids:
        try:
            match_id_int = int(match_id)
            eventos = sb.events(match_id=match_id_int)
            eventos['match_id'] = match_id
            eventos_todos.append(eventos)
        except Exception as e:
            print(f"❌ Error en partido {match_id}: {e}")
            continue

    return pd.concat(eventos_todos, ignore_index=True) if eventos_todos else pd.DataFrame()

datos = laliga_16

match_ids = datos['match_id'].tolist()

df_eventos = obtener_eventos_por_partidos(match_ids)
```

```python
import os
import time
import pandas as pd
from statsbombpy import sb

def guardar_eventos_por_partido(match_ids, carpeta_salida="eventos_guardados"):
    os.makedirs(carpeta_salida, exist_ok=True)
    guardados = 0

    for i, match_id in enumerate(match_ids, 1):
        ruta_archivo = os.path.join(carpeta_salida, f"match_{match_id}.pkl")

        if os.path.exists(ruta_archivo):
            continue  # Ya descargado

        try:
            eventos = sb.events(match_id=int(match_id))
            eventos['match_id'] = match_id
            eventos.to_pickle(ruta_archivo)
            guardados += 1
        except Exception as e:
            print(f"❌ Error con partido {match_id}: {e}")
            continue

        if i % 20 == 0 or i == len(match_ids):
            total_guardados = len(os.listdir(carpeta_salida))
            print(f"📦 Guardados {total_guardados} partidos hasta ahora")

        time.sleep(1)  # Espera para no saturar

    print(f"\n✅ Proceso terminado. Total partidos guardados: {len(os.listdir(carpeta_salida))}")


```

```python
import pandas as pd

def calcular_PSxG_por_portero(df_eventos):
    """
    Calcula el PSxG (post-shot xG) enfrentado por cada portero.
    Suma el xG de los tiros que han ido a portería: Goal, Saved, Saved To Post.
    Asigna ese xG al portero del equipo rival (evento de tipo 'Goal Keeper').
    """
    # 1. Filtrar tiros a portería
    tiros = df_eventos[
        (df_eventos['type'] == 'Shot') &
        (df_eventos['shot_outcome'].isin(['Goal', 'Saved', 'Saved To Post']))
    ][['match_id', 'team', 'shot_statsbomb_xg']].copy()

    # 2. Filtrar eventos de portero (evento: 'Goal Keeper')
    porteros = df_eventos[df_eventos['type'] == 'Goal Keeper']
    porteros = porteros[['match_id', 'player', 'team']].drop_duplicates()

    # 3. Asociar cada tiro con el portero rival del partido (no del mismo equipo que tiró)
    psxg_data = []
    for _, tiro in tiros.iterrows():
        match_id = tiro['match_id']
        equipo_tirador = tiro['team']
        xg = tiro['shot_statsbomb_xg']

        # Buscar portero del equipo rival
        porteros_partido = porteros[(porteros['match_id'] == match_id) & (porteros['team'] != equipo_tirador)]

        if not porteros_partido.empty:
            portero = porteros_partido.iloc[0]['player']
            psxg_data.append({'player': portero, 'PSxG': xg})

    # 4. Agrupar por portero
    df_psxg = pd.DataFrame(psxg_data)
    df_psxg = df_psxg.groupby('player').sum().reset_index()

    return df_psxg

def calcular_OPA_por_portero(df_eventos):
    """
    OPA: Número de acciones defensivas realizadas fuera del área (tipo 'Pressure', 'Ball Recovery', 'Interception', 'Duel')
    por parte del portero, considerando que se hacen fuera del área (x < 102 o y fuera de [18, 62])
    """
    acciones_defensivas = df_eventos[
        (df_eventos['player'].notnull()) &
        (df_eventos['type'].isin(['Pressure', 'Ball Recovery', 'Interception', 'Duel']))
    ].copy()

    # Porteros que hacen acciones fuera del área (location no dentro del área)
    def fuera_del_area(loc):
        if isinstance(loc, list) and len(loc) == 2:
            x, y = loc
            return x < 102 or y < 18 or y > 62
        return False

    acciones_defensivas = acciones_defensivas[acciones_defensivas['location'].apply(fuera_del_area)]

    return acciones_defensivas.groupby('player').size().reset_index(name='OPA')

def porterias_a_cero_por_portero(df_eventos, df_jugadores, df_partidos):
    """
    Calcula cuántas porterías a cero (imbatibilidades) ha logrado cada portero
    en los partidos jugados.

    Parámetros:
    - df_eventos: DataFrame con eventos de partidos.
    - df_jugadores: DataFrame con los jugadores (debe tener columnas 'player' y 'player_id').
    - df_partidos: DataFrame con información de los partidos (match_id, home_team, away_team, home_score, away_score).

    Retorna:
    - DataFrame con player_id, player, imbatibilidades
    """
    from collections import defaultdict

    imbatibilidades = defaultdict(int)

    match_ids = df_partidos['match_id'].unique()

    for match_id in match_ids:
        try:
            eventos_partido = df_eventos[df_eventos['match_id'] == match_id]
            alineaciones = eventos_partido[eventos_partido['type'] == 'Starting XI']

            # Extraer equipos y goles
            partido_info = df_partidos[df_partidos['match_id'] == match_id].iloc[0]
            home_team = partido_info['home_team']
            away_team = partido_info['away_team']
            home_goals = partido_info['home_score']
            away_goals = partido_info['away_score']

            for _, row in alineaciones.iterrows():
                equipo = row['team']
                es_local = equipo == home_team
                goles_en_contra = away_goals if es_local else home_goals

                # Solo si dejó portería a cero
                if goles_en_contra == 0:
                    lineup = row.get('tactics', {}).get('lineup', [])
                    for jugador in lineup:
                        if isinstance(jugador, dict):
                            if jugador['position']['name'] == 'Goalkeeper':
                                pid = jugador['player']['id']
                                imbatibilidades[pid] += 1

        except Exception as e:
            print(f"❌ Error en partido {match_id}: {e}")
            continue

    # Convertimos a DataFrame
    df_imbatible = pd.DataFrame([
        {'player_id': pid, 'imbatibilidades': count}
        for pid, count in imbatibilidades.items()
    ])

    # Unimos con los nombres
    df_resultado = df_jugadores.merge(df_imbatible, on='player_id', how='left')
    df_resultado['imbatibilidades'] = df_resultado['imbatibilidades'].fillna(0).astype(int)

    return df_resultado
```
