```python
import pandas as pd
from datasets import Dataset
from ragas import evaluate
from ragas.metrics import answer_relevancy, answer_correctness

```

```python
# Cargar archivos
agent_df = pd.read_csv("evaluacion_agente.csv")
gt_df = pd.read_csv("ground_truth_actualizado.csv")

# Asegúrate de que los nombres de columna coincidan (puedes adaptarlos aquí si no)
# Suponemos que las columnas se llaman: question, ground_truth, prediction
df = pd.DataFrame({
    "question": agent_df["question"],
    "ground_truth": gt_df["expected_players"],
    "answer": agent_df["agent_answer"]
})

# Convertir a HuggingFace Dataset
ragas_ds = Dataset.from_pandas(df)
```

```python
from ragas import evaluate
from ragas.metrics import answer_correctness

# Ejecutar solo con métricas que no requieran contexto
result = evaluate(
    ragas_ds,
    metrics=[answer_correctness]
)

# Mostrar resultados
result.to_pandas()
```
