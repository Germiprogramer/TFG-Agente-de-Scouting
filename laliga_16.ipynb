```python
laliga_16 = sb.matches(competition_id=11, season_id=27)
bundesliga_16 = sb.matches(competition_id=9, season_id=27)
premierleague_16 = sb.matches(competition_id=2, season_id=27)
seriea_16 = sb.matches(competition_id=12, season_id=27)
ligue1_16 = sb.matches(competition_id=7, season_id=27)

ligas_16 = pd.concat([laliga_16, bundesliga_16, premierleague_16, seriea_16, ligue1_16])
ligas_16
```

```python
# Lista para guardar DataFrames de eventos
eventos_lista = []

# Descargar eventos de cada partido
for match_id in seriea_16['match_id']:

        eventos = sb.events(match_id=match_id)
        eventos['match_id'] = match_id  # Añadir identificador de partido
        eventos_lista.append(eventos)


# Unir todos los DataFrames en uno solo
df_eventos = pd.concat(eventos_lista, ignore_index=True)

# Guardar como archivo CSV
#df_eventos.to_csv('eventos_seriea_2016.csv', index=False)
```
