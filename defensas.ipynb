```python
def resumen_duelos_agrupado(df_eventos, porcentaje=True):
    """
    Agrupa los resultados de duelos en 'Ganado' y 'Perdido' por tipo de duelo.
    
    Parámetros:
    - df_eventos: DataFrame con eventos
    - porcentaje: si True, devuelve %; si False, conteo bruto

    Retorna:
    - DataFrame con tipos de duelo y porcentaje (o conteo) de ganados/perdidos
    """
    df = df_eventos.copy()

    # Extraer nombres de tipo de duelo y resultado
    df['duel_type_name'] = df['duel_type'].apply(lambda x: x['name'] if isinstance(x, dict) else x)
    df['duel_outcome_name'] = df['duel_outcome'].apply(lambda x: x['name'] if isinstance(x, dict) else x)

    # Eliminar filas vacías
    df = df.dropna(subset=['duel_type_name', 'duel_outcome_name'])

    # Clasificar resultados
    ganados = ['Won', 'Success In Play', 'Success Out']
    df['resultado_duelo'] = df['duel_outcome_name'].apply(lambda x: 'Ganado' if x in ganados else 'Perdido')

    # Tabla agrupada
    tabla = pd.crosstab(df['duel_type_name'], df['resultado_duelo'])

    if porcentaje:
        tabla = tabla.div(tabla.sum(axis=1), axis=0).round(3) * 100  # convertir a %

    return tabla

resumen_duelos_agrupado(mascherano_duelos)
```

```python
def valores_unicos_duel_outcome_aerial(df_eventos):
    """
    Devuelve los valores únicos de 'duel_outcome' para duelos de tipo 'Aerial Duel'.
    
    Parámetro:
    - df_eventos: DataFrame de eventos

    Retorna:
    - Lista de valores únicos en duel_outcome para Aerial Duel
    """
    df = df_eventos.copy()

    # Extraer nombres limpios
    df['duel_type_name'] = df['duel_type'].apply(lambda x: x['name'] if isinstance(x, dict) else x)
    df['duel_outcome_name'] = df['duel_outcome'].apply(lambda x: x['name'] if isinstance(x, dict) else x)

    # Filtrar por duelos aéreos
    df_aerial = df[df['duel_type_name'] == 'Aerial Duel']

    return df_aerial['duel_outcome_name'].dropna().unique()

valores_unicos_duel_outcome_aerial(mascherano)
```

```python
def intercepciones_exitosas_totales(df_eventos):
    """
    Devuelve el número total de intercepciones exitosas (Won + Success In Play).

    Retorna:
    - Entero: número de intercepciones exitosas
    """
    df = df_eventos.copy()
    df = df.dropna(subset=['interception_outcome'])

    df['interception_outcome'] = df['interception_outcome'].apply(lambda x: x['name'] if isinstance(x, dict) else x)

    exitosas = ['Won', 'Success In Play']
    return df['interception_outcome'].isin(exitosas).sum()


intercepciones_exitosas_totales(busquets)
```

```python
def duelos_aereos_ganados(df):
    """Número total de duelos aéreos ganados (vía despejes + tiros)."""
    despejes = df['clearance_aerial_won'].sum() if 'clearance_aerial_won' in df.columns else 0
    tiros = df['shot_aerial_won'].sum() if 'shot_aerial_won' in df.columns else 0
    return despejes + tiros

def porcentaje_duelos_aereos_ganados(df):
    """Porcentaje total de duelos aéreos ganados respecto al total (ganados + perdidos)."""
    ganados = duelos_aereos_ganados(df)
    perdidos = df['duel_type'].eq('Aerial Lost').sum() if 'duel_type' in df.columns else 0
    total = ganados + perdidos
    return 100 * ganados / total if total > 0 else 0


def duelos_aereos_ganados_despeje(df):
    """Número de despejes con la cabeza ganados como duelo aéreo."""
    if 'clearance_aerial_won' in df.columns and 'clearance_head' in df.columns:
        return df[(df['clearance_aerial_won'] == True) & (df['clearance_head'] == True)].shape[0]
    return 0

def porcentaje_duelos_aereos_ganados_despeje(df):
    """Porcentaje de despejes con la cabeza que ganaron un duelo aéreo."""
    if 'clearance_head' not in df.columns:
        return 0
    total_cabeza = df[df['clearance_head'] == True].shape[0]
    ganados = duelos_aereos_ganados_despeje(df)
    return 100 * ganados / total_cabeza if total_cabeza > 0 else 0


def duelos_aereos_ganados_tiro(df):
    """Número de tiros de cabeza que ganaron un duelo aéreo."""
    if 'shot_aerial_won' in df.columns and 'shot_body_part' in df.columns:
        return df[(df['shot_aerial_won'] == True) & (df['shot_body_part'] == 'Head')].shape[0]
    return 0

def porcentaje_duelos_aereos_ganados_tiro(df):
    """Porcentaje de tiros de cabeza que ganaron un duelo aéreo."""
    if 'shot_body_part' not in df.columns:
        return 0
    total_cabezazos = df[df['shot_body_part'] == 'Head'].shape[0]
    ganados = duelos_aereos_ganados_tiro(df)
    return 100 * ganados / total_cabezazos if total_cabezazos > 0 else 0


#comprobar todo mañana
```
