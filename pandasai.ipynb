```python
import pandas as pd
from pandasai import SmartDataframe
from langchain_community.llms import Ollama
```

```python
prefix = """
Eres un agente de inteligencia artificial especializado en fútbol. 
Tu misión es ayudar a evaluar jugadores basándote exclusivamente en los datos del DataFrame.
No inventes respuestas. Si no hay suficiente información, dilo claramente.
"""
```

```python
# ⚠️ Cambia el modelo si usas otro (como llama2, gemma, etc.)
llm = Ollama(model="codellama:7b-instruct")
```

```python
# Carga tu DataFrame
df = pd.read_csv("datos/datos_jugadores_v4/jugadores_total.csv")
new_df = df.copy()
```

```python
# Crea un SmartDataframe con el modelo
sdf = SmartDataframe(new_df, config={"llm": llm, "custom_head": """
        Eres un agente experto en scouting futbolístico. 
        Solo puedes usar los datos disponibles en el DataFrame para responder.
        Si no hay suficiente información, indica que no puedes responder.
        """,})
```

```python
# Hazle una pregunta al agente en lenguaje natural
respuesta = sdf.chat("¿Quién es el jugador con rating alto?")
print("🤖", respuesta)
```
