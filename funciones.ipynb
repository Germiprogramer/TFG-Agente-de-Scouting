```python
def obtener_eventos_por_partidos(lista_match_ids):
    """
    Descarga y concatena todos los eventos de los partidos dados.
    """
    eventos_todos = []

    for match_id in lista_match_ids:
        try:
            match_id_int = int(match_id)
            eventos = sb.events(match_id=match_id_int)
            eventos['match_id'] = match_id
            eventos_todos.append(eventos)
        except Exception as e:
            print(f"❌ Error en partido {match_id}: {e}")
            continue

    return pd.concat(eventos_todos, ignore_index=True) if eventos_todos else pd.DataFrame()

match_ids_barça = fc_barcelona['match_id'].tolist()

eventos_barça = obtener_eventos_por_partidos(match_ids_barça)
```

```python
def resumen_duelos_agrupado(df_eventos, porcentaje=True):
    """
    Agrupa los resultados de duelos en 'Ganado' y 'Perdido' por tipo de duelo.
    
    Parámetros:
    - df_eventos: DataFrame con eventos
    - porcentaje: si True, devuelve %; si False, conteo bruto

    Retorna:
    - DataFrame con tipos de duelo y porcentaje (o conteo) de ganados/perdidos
    """
    df = df_eventos.copy()

    # Extraer nombres de tipo de duelo y resultado
    df['duel_type_name'] = df['duel_type'].apply(lambda x: x['name'] if isinstance(x, dict) else x)
    df['duel_outcome_name'] = df['duel_outcome'].apply(lambda x: x['name'] if isinstance(x, dict) else x)

    # Eliminar filas vacías
    df = df.dropna(subset=['duel_type_name', 'duel_outcome_name'])

    # Clasificar resultados
    ganados = ['Won', 'Success In Play', 'Success Out']
    df['resultado_duelo'] = df['duel_outcome_name'].apply(lambda x: 'Ganado' if x in ganados else 'Perdido')

    # Tabla agrupada
    tabla = pd.crosstab(df['duel_type_name'], df['resultado_duelo'])

    if porcentaje:
        tabla = tabla.div(tabla.sum(axis=1), axis=0).round(3) * 100  # convertir a %

    return tabla

resumen_duelos_agrupado(mascherano_duelos)
```

```python
def contar_ball_recoveries(df_eventos):
    """
    Devuelve un DataFrame con el número de recuperaciones de balón por jugador.
    """
    # Filtrar eventos donde existe alguna recuperación (ofensiva o no fallida)
    df = df_eventos[df_eventos['ball_recovery_offensive'].notna() | df_eventos['ball_recovery_recovery_failure'].notna()]

    # Filtrar eventos sin jugador
    df_s = df.dropna(subset=['player_id', 'player'])

    # Agrupar y contar recuperaciones por jugador
    recuperaciones = df_s.groupby(['player_id', 'player']).size().reset_index(name='ball_recoveries')

    return recuperaciones

contar_ball_recoveries(messi)
```

```python
def filtrar_pases_completados(df):
    """
    Devuelve solo los pases que fueron completados (pass_outcome == NaN).
    """
    df_filtrado = df[~df['pass_outcome'].isin(['Unknown', 'Injury Clearance'])]
    return df_filtrado[df_filtrado['pass_outcome'].isnull()]

```

```python
def total_pases(df):
    """
    Número total de pases válidos.
    Excluye 'Unknown' y 'Injury Clearance'.
    """
    return df[~df['pass_outcome'].isin(['Unknown', 'Injury Clearance'])].shape[0]

def pases_completados(df):
    """
    Número de pases completados.
    Son aquellos sin 'pass_outcome' (i.e. NaN).
    """
    df_filtrado = df[~df['pass_outcome'].isin(['Unknown', 'Injury Clearance'])]
    return df_filtrado['pass_outcome'].isnull().sum()

def pases_incompletos(df):
    """
    Número de pases incompletos ('Incomplete').
    """
    return df[df['pass_outcome'] == 'Incomplete'].shape[0]

def pases_fuera(df):
    """
    Número de pases que acabaron fuera ('Out').
    """
    return df[df['pass_outcome'] == 'Out'].shape[0]

def pases_fuera_de_juego(df):
    """
    Número de pases completados pero anulados por fuera de juego ('Pass Offside').
    """
    return df[df['pass_outcome'] == 'Pass Offside'].shape[0]

def pases_fallidos(df):
    """
    Suma total de pases fallidos (Incomplete + Out + Pass Offside).
    """
    return (
        pases_incompletos(df) +
        pases_fuera(df) +
        pases_fuera_de_juego(df)
    )

def porcentaje_pases_completados(df):
    """
    Porcentaje de pases completados respecto al total válido.
    """
    total = total_pases(df)
    completados = pases_completados(df)
    return 100 * completados / total if total > 0 else 0

```

```python
def media_longitud_pase(df):
    """Promedio de longitud de los pases COMPLETADOS."""
    # Filtrar solo los pases completados
    df = filtrar_pases_completados(df)
    return df['pass_length'].mean()

def media_angulo_pase(df):
    """Promedio del ángulo del pase (en radianes). 0 es recto hacia adelante, π es hacia atrás. Se miden solo los pases completados."""
    # Filtrar solo los pases completados
    df = filtrar_pases_completados(df)
    return df['pass_angle'].mean()
```

```python
def total_pases_altura(df, altura):
    """
    Número de pases con una altura específica ('Ground Pass', 'Low Pass', 'High Pass').
    Ignora los valores NaN en pass_height.
    """
    return df[df['pass_height'] == altura].shape[0]

def porcentaje_pases_altura(df, altura):
    """
    Porcentaje de pases con cierta altura sobre los que tienen altura conocida.
    No incluye NaN en pass_height.
    """
    df_validos = df[df['pass_height'].notnull()]
    total_con_altura = df_validos.shape[0]
    cantidad = df_validos[df_validos['pass_height'] == altura].shape[0]
    return 100 * cantidad / total_con_altura if total_con_altura > 0 else 0

```

```python
def total_crosses(df):
    """Número de centros (crosses) realizados."""
    return df[df['pass_cross'] == True].shape[0]

def total_crosses_completados(df):
    df = filtrar_pases_completados(df)
    return df[df['pass_cross'] == True].shape[0]

def total_cutbacks(df):
    """Número de pases tipo cut-back (atrás en el área)."""
    return df[df['pass_cut_back'] == True].shape[0]

def total_cutbacks_completados(df):
    df = filtrar_pases_completados(df)
    return df[df['pass_cut_back'] == True].shape[0]

def total_switches(df):
    """Número de cambios de orientación (switches)."""
    return df[df['pass_switch'] == True].shape[0]

def total_switches_completados(df):
    df = filtrar_pases_completados(df)
    return df[df['pass_switch'] == True].shape[0]

def total_miscommunication(df):
    """Número de pases fallados por malentendido (no error técnico)."""
    return df[df['pass_miscommunication'] == True].shape[0]

def total_deflected(df):
    """Número de pases que fueron desviados (deflected)."""
    return df[df['pass_deflected'] == True].shape[0]
```

```python
def total_pases_con_body_part(df):
    """
    Número total de pases que tienen body_part definido (no NaN).
    """
    return df['pass_body_part'].notnull().sum()

def porcentaje_pases_cabeza(df):
    """
    Porcentaje de pases realizados con la cabeza (sobre los que tienen body_part definido).
    """
    total = total_pases_con_body_part(df)
    cabeza = df[df['pass_body_part'] == 'Head'].shape[0]
    return 100 * cabeza / total if total > 0 else 0

def porcentaje_pases_pie_derecho(df):
    """
    Porcentaje de pases realizados con el pie derecho.
    """
    total = total_pases_con_body_part(df)
    derecho = df[df['pass_body_part'] == 'Right Foot'].shape[0]
    return 100 * derecho / total if total > 0 else 0

def porcentaje_pases_pie_derecho_acertados(df):
    """
    Porcentaje de pases con el pie derecho que fueron completados.
    """
    derecho_df = df[df['pass_body_part'] == 'Right Foot']  # esto es un DataFrame
    total = derecho_df.shape[0]
    completados = filtrar_pases_completados(derecho_df).shape[0]
    
    return 100 * completados / total if total > 0 else 0

def porcentaje_pases_pie_izquierdo(df):
    """
    Porcentaje de pases realizados con el pie izquierdo.
    """
    total = total_pases_con_body_part(df)
    izquierdo = df[df['pass_body_part'] == 'Left Foot'].shape[0]
    return 100 * izquierdo / total if total > 0 else 0

def porcentaje_pases_pie_izquierdo_acertados(df):
    """
    Porcentaje de pases con el pie izquierdo que fueron completados.
    """
    izquierdo_df = df[df['pass_body_part'] == 'Left Foot']  # esto es un DataFrame
    total = izquierdo_df.shape[0]
    completados = filtrar_pases_completados(izquierdo_df).shape[0]
    
    return 100 * completados / total if total > 0 else 0
```

```python
def pases_en_campo_propio(df):
    """
    Número de pases que comienzan en campo propio (x < 60).
    """
    new_df = df[~df['pass_outcome'].isin(['Unknown', 'Injury Clearance'])]
    """
    Número de pases que comienzan en campo propio (x < 60).
    """
    return new_df[new_df['location'].apply(lambda loc: isinstance(loc, list) and loc[0] < 60)].shape[0]


def pases_en_campo_rival(df):
    """
    Número de pases que comienzan en campo rival (x >= 60).
    """
    new_df = df[~df['pass_outcome'].isin(['Unknown', 'Injury Clearance'])]

    return new_df[new_df['location'].apply(lambda loc: isinstance(loc, list) and loc[0] >= 60)].shape[0]

```

```python
def porcentaje_pases_en_campo_rival(df):
    """
    Porcentaje de pases válidos que comienzan en campo rival (x >= 60),
    excluyendo los pases con outcome 'Unknown' o 'Injury Clearance'.
    """
    df_filtrado = df[~df['pass_outcome'].isin(['Unknown', 'Injury Clearance'])]
    total = df_filtrado.shape[0]
    en_rival = df_filtrado[df_filtrado['location'].apply(lambda loc: isinstance(loc, list) and loc[0] >= 60)].shape[0]
    return 100 * en_rival / total if total > 0 else 0
```

```python
import matplotlib.pyplot as plt

def dibujar_campo(ax, largo=120, ancho=80):
    # Dibujar límites del campo
    ax.plot([0, largo], [0, 0], color="black")
    ax.plot([0, largo], [ancho, ancho], color="black")
    ax.plot([0, 0], [0, ancho], color="black")
    ax.plot([largo, largo], [0, ancho], color="black")

    # Dibujar líneas medias y áreas
    ax.plot([60, 60], [0, ancho], color="black", linestyle="--")  # línea del medio
    ax.add_patch(plt.Rectangle((102, 18), 18, 44, fill=False))     # área derecha
    ax.add_patch(plt.Rectangle((0, 18), 18, 44, fill=False))       # área izquierda
    ax.add_patch(plt.Circle((60, 40), 10, fill=False))             # círculo central
    ax.set_xlim(0, largo)
    ax.set_ylim(0, ancho)
    ax.set_aspect('equal')
    ax.axis('off')

def graficar_pases(df):
    fig, ax = plt.subplots(figsize=(12, 8))
    dibujar_campo(ax)

    for _, row in df.iterrows():
        start = row['location']
        end = row['pass_end_location']
        if isinstance(start, list) and isinstance(end, list):
            ax.arrow(
                start[0], start[1],
                end[0] - start[0], end[1] - start[1],
                head_width=1.5, head_length=2, fc='blue', ec='blue', alpha=0.7
            )

    ax.set_title("Primeros 50 pases de bravo", fontsize=16)
    plt.show()
```

```python
import numpy as np

def es_progresivo_wyscout(start, end):
    """
    Determina si un pase es progresivo según criterios tipo Wyscout.
    """
    if not (isinstance(start, list) and isinstance(end, list)):
        return False

    # Distancia al gol rival en eje X (portería en x = 120)
    inicio = 120 - start[0]
    fin = 120 - end[0]
    ganancia = inicio - fin  # cuánto se avanza hacia el gol

    # Clasificación por zona de campo
    en_propio = start[0] < 60 and end[0] < 60
    cambia_de_mitad = start[0] < 60 and end[0] >= 60
    en_rival = start[0] >= 60 and end[0] >= 60

    # Reglas
    if en_propio and ganancia >= 30:
        return True
    elif cambia_de_mitad and ganancia >= 15:
        return True
    elif en_rival and ganancia >= 10:
        return True
    else:
        return False
```

```python
def contar_pases_progresivos(df):
    return df[df.apply(lambda row: es_progresivo_wyscout(row['location'], row['pass_end_location']), axis=1)].shape[0]

def contar_pases_progresivos_completados(df):
    df = filtrar_pases_completados(df)
    return df[df.apply(lambda row: es_progresivo_wyscout(row['location'], row['pass_end_location']), axis=1)].shape[0]

def porcentaje_pases_progresivos_completados(df):
    """
    Porcentaje de pases progresivos respecto al número de pases progresivos.
    """
    total_progresivos = contar_pases_progresivos(df)
    progresivos_completados = contar_pases_progresivos_completados(df)
    return 100 * progresivos_completados / total_progresivos if total_progresivos > 0 else 0
```

```python
def precision_pases_progresivos(df):
    """
    Calcula el porcentaje de pases progresivos completados.
    Solo cuenta como exitosos los que no tienen 'pass_outcome' (NaN).
    """
    df_prog = df[df.apply(lambda row: es_progresivo_wyscout(row['location'], row['pass_end_location']), axis=1)]
    if df_prog.empty:
        return 0
    completados = df_prog['pass_outcome'].isnull().sum()
    return 100 * completados / len(df_prog)
```
